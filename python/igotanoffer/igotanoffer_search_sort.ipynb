```python
"""
https://leetcode.com/problems/k-closest-points-to-origin/
We have a list of points on the plane.  Find the K closest points to the origin (0, 0).

Given an array of points where points[i] = [xi, yi] represents a point on the X-Y plane and an integer k,
return the k closest points to the origin (0, 0).

The distance between two points on the X-Y plane is the Euclidean distance (i.e., âˆš(x1 - x2)2 + (y1 - y2)2).

You may return the answer in any order. The answer is guaranteed to be unique (except for the order that it is in).
"""

import heapq

def findKClosestPoints(points, k):
    # finding top k is easist with a priority queue
    # python heapq is a min heap
    heap = []
    # heapq.heappush()
    # heapq.heappop()

    def _distToOrigin(a,b):
        return pow( a*a + b*b , 0.5)

    for p in points:
        d = _distToOrigin(p[0], p[1])
        heapq.heappush(heap, (d, p))

    retVal = []
    for _ in range(k):
        d, p = heapq.heappop(heap)
        retVal.append(p)

    return retVal

tests = [
([], 0, []),
([[1,3],[-2,2]], 2, [[-2,2],[1,3]]),
([[1,3],[-2,2]], 0, []),
([[1,3],[-2,2]], 1, [[-2,2]]),
([[3,3],[5,-1],[-2,4]], 2, [[3,3],[-2,4]])
]
for test in tests:
    assert(sorted(findKClosestPoints(test[0], test[1])) == sorted(test[2]))
```

```python
"""
https://leetcode.com/problems/intersection-of-two-arrays/
Given two arrays, write a function to compute their intersection.
"""

def getIntersection(a, b):
    # # set
    # return list(set(a).intersection(set(b)))

    retVal = []
    aSet = set(a)
    for e in set(b):
        if e in aSet:
            retVal.append(e)
    return retVal

tests = [
([1,2,2,1], [2,2], [2]),
([4,9,5], [9,4,9,8,4], [4,9])
]
for test in tests:
    assert(sorted(getIntersection(test[0],test[1])) == sorted(test[2]))
```
