```python
"""
https://leetcode.com/problems/two-sum-bsts/
https://leetcode.ca/2019-03-28-1214-Two-Sum-BSTs/


Given the roots of two binary search trees, root1 and root2, 
return true if and only if 
there is a node in the first tree and a node in the second tree whose values sum up to a given integer target.
 
Constraints:
The number of nodes in each tree is in the range [1, 5000].
-10^9 <= Node.val, target <= 10^9
"""

class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def readTree(vals):
    nodeMap={}
    for i in range(len(vals)-1, -1, -1):
        nodeMap[i]=Node(vals[i])
        nodeMap[i].left = nodeMap[i*2+1] if i*2+1 in nodeMap else None
        nodeMap[i].right = nodeMap[i*2+2] if i*2+2 in nodeMap else None
    return nodeMap[0]


# 1. for each elem in root1 (N), search root2 (M) of (target-elem) exists in root2: Time smaller of O(N * log(M)) or O(M *log(N)); Space O(1)
# 2. get all elements of root1 and root2 in array (in-order; sorted), and convert the problem to two sorted arrays: time O(N+M); space O(N+M)
def twoSumBSTs(root1, root2, target):
    elem1=[]
    elem2=[]
    def _traverseInorder(node, elem):
        if node == None:
            return
        _traverseInorder(node.left, elem)
        if node.val != None:
            elem.append(node.val)
        _traverseInorder(node.right, elem)
    _traverseInorder(root1, elem1)
    _traverseInorder(root2, elem2)
    # print("elem1", elem1)
    # print("elem2", elem2)

    p1 = 0 # for elem1
    p2 = len(elem2)-1 # for elem2

    while p1 < len(elem1) and 0 <= p2:
        diff = target - (elem1[p1] + elem2[p2])
        if diff == 0:
            return True
        if diff > 0:
            p2 -= 1
        else:
            p1 += 1

    return False


tests = [
    ([2,1,4], [1,0,3], 5, True),
    #2 and 3 sum up to 5
    ([0,-10,10], [5,1,7,0,2], 18, False),
    ([0,-10,10], [5,1,7,None,2,None,8], 18, True)
]
for t in tests:
    root1=readTree(t[0])
    root2=readTree(t[1])
    twoSumBSTs(root1, root2, t[2])
    #assert(twoSumBSTs(root1, root2, t[2]) == t[3])
```
