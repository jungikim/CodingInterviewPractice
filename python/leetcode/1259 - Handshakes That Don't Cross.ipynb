```python
"""
https://leetcode.com/problems/handshakes-that-dont-cross/
https://leetcode.ca/2019-05-12-1259-Handshakes-That-Don't-Cross/

You are given an even number of people numPeople that stand around a circle and 
each person shakes hands with someone else so that there are numPeople / 2 handshakes total.

Return the number of ways these handshakes could occur such that none of the handshakes cross.

Since the answer could be very large, return it modulo 10^9 + 7.


Constraints:
2 <= numPeople <= 1000
numPeople is even.
"""

def handshakesNoCross(numPeople):
    numShakesMap = { # people: numShakes
        0: 1,
        2: 1,
        4: 2,
        6: 5
    }
    def _numShakes(n):
        if n in numShakesMap:
            return numShakesMap[n]

        numShakes = 0
        i = 1
        for j in range(i+1, n+1, 2):
            # i-j handshake partitions the rest of the people into two
            #  p1: i+1 to j-1: can be 0, 2, ...
            #  p2: j+1 ... n 1 .. i-1  or n - p1 - 2

            p1 = (j-1) - (i+1) + 1
            p2 = n - p1 - 2
            n1 = _numShakes(p1)
            n2 = _numShakes(p2)
            
            # print(f"{i}-{j}: _numShakes({p1}) * _numShakes({p2}) = {n1} * {n2}")
            numShakes += n1 * n2

        numShakesMap[n] = numShakes
        # print(f"numShakesMap[{n}] = {numShakes}")
        return numShakes

    return _numShakes(numPeople)


tests = [
    (2, 1),
    (4, 2),
    #  1   2
    #  4   3
    #There are two ways to do it, the first way is [(1,2),(4,3)] and the second one is [(2,3),(4,1)].
    (6, 5),
    #     2
    #  1     3
    #  6     4
    #     5
    # 1-2 3-4 5-6
    # 1-2 3-6 4-5

    # 1-4 2-3 5-6

    # 1-6 2-3 4-5
    # 1-6 2-5 3-4

    (8, 14),
    #     1   2
    #   8        3
    #
    #   7        4
    #     6   5
    #
    # 12 34 56 78
    # 12 34 58 67
    # 12 36 45 78
    # 12 38 45 67
    # 12 38 47 56

    # 14 23 56 78
    # 14 23 58 67

    # 16 23 45 78
    # 16 25 34 78

    # 18 23 45 67
    # 18 25 34 67
    # 18 27 34 56
    # 18 27 36 45
    
    
]
for t in tests:
    retVal = handshakesNoCross(t[0])
    print(t, retVal)
    assert(retVal == t[1])
```
