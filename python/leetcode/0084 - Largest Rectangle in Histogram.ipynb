```python
"""
Given an array of integers heights representing the histogram's bar height where the width of each bar is 1,
return the area of the largest rectangle in the histogram.

Constraints:
1 <= heights.length <= 105
0 <= heights[i] <= 104

monotonic stack
"""

def largestRectangleArea(heights):
    if len(heights)==0:
        return 0
    if len(heights)==1:
        return heights[0]

    n = len(heights)
    stack = []
    left = [-1] * n # stores the index of the left-most bar shorter than current bar
    right = [n] * n # stores the index of the right-most bar shorter than current bar
    for i, h in enumerate(heights):
        print(f"idx {i}, height {h}")
        while stack and heights[stack[-1]] >= h:
            print(f"Update right[{stack[-1]}] as idx {i} (height of {stack[-1]} ({heights[stack[-1]]}) >= {h})")
            right[stack[-1]] = i
            print(f"Pop {stack[-1]}")
            stack.pop()
        if stack:
            print(f"Update left[{i}] as {stack[-1]}")
            left[i] = stack[-1]
        print(f"Append {i}")
        stack.append(i)

    print("final status")
    print(f"{heights = }")
    print(f"{left = }")
    print(f"{right = }")
    for i, h in enumerate(heights):
        area = h * (right[i] - left[i] - 1)
        print(i, h, area)
    return max(h * (right[i] - left[i] - 1) for i, h in enumerate(heights))


tests = [
    ([2,1,5,6,2,3], 10),
    # put 0 in stack
    # 

    
    ([2,4], 4),
]
for t in tests:
    assert(largestRectangleArea(t[0]) == t[1])
```
