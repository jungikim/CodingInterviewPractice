```python
"""
https://leetcode.com/problems/median-of-two-sorted-arrays/

Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

 
Constraints:
nums1.length == m
nums2.length == n
0 <= m <= 1000
0 <= n <= 1000
1 <= m + n <= 2000
-10^6 <= nums1[i], nums2[i] <= 10^6
"""

from typing import List
import heapq
def findMedianSortedArrays(nums1: List[int], nums2: List[int]) -> float:
    # approach 1:
    #   binary search on the two lists
    #
    # approach 2:
    # we maintain two heaps
    #   one maxheap for lower half of numbers
    #   one minheap for upper half of numbers
    # len(lowerHeap) >= len(upperHeap)
    # add(): insert to upperHeap, while len(lowerHeap) < len(upperHeap): pop from upperHeap and insert to lowerHeap

    lowerHeap = [] # maxheap (-val)
    upperHeap = [] # minheap (val)

    m = len(nums1)
    n = len(nums2)
    mn = m+n

    def _add(v):
        heapq.heappush(upperHeap, v)
        while len(lowerHeap) < len(upperHeap):
            _v = heapq.heappop(upperHeap)
            heapq.heappush(lowerHeap, -_v)
    for i in range(m):
        _add(nums1[i])
    for i in range(n):
        _add(nums2[i])

    if mn %2 == 1:
        return -lowerHeap[0] * 1.0

    return (-lowerHeap[0] + upperHeap[0]) / 2.0

tests = [
([1,3], [2], 2.00000),   
# [1,2,3] and median is 2.
([1,2], [3,4], 2.50000),    
# [1,2,3,4] and median is (2 + 3) / 2 = 2.5 
]

for t in tests:
    retVal = findMedianSortedArrays(t[0], t[1])
    assert(retVal == t[2])
```
