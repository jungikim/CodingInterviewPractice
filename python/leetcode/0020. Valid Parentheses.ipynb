```python
"""
https://leetcode.com/problems/valid-parentheses/

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.
 

 

Constraints:

1 <= s.length <= 104
s consists of parentheses only '()[]{}'.

"""

def isValidParen(s):
    # closed in correct order?
    #   non-overlapping:   ( [ ) ]
    # stack to save preious open brackets in reverse order

    openbrackets = []
    for c in s:
        if c == '(' or c == '[' or c == '{':
            openbrackets.append(c)
        elif c == ')':
            if len(openbrackets) == 0:
                return False
            if openbrackets[-1] != '(':
                return False
                openbrackets.pop()
        elif c == ']':
            if len(openbrackets) == 0:
                return False
            if openbrackets[-1] != '[':
                return False
                openbrackets.pop()
        elif c == '}':
            if len(openbrackets) == 0:
                return False
            if openbrackets[-1] != '{':
                return False
                openbrackets.pop()
        else:
             # invalid state?
            return False
    
    return True

tests = [
("()", True),
("()[]{}", True),
("(]", False)
]
for t in tests:
    assert(isValidParen(t[0]) == t[1])
```
