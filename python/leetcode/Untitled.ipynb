```python
"""
https://leetcode.com/problems/search-in-rotated-sorted-array/


There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly rotated at an unknown pivot index k (1 <= k < nums.length)
such that the resulting array is [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] (0-indexed).
For example, [0,1,2,4,5,6,7] might be rotated at pivot index 3 and become [4,5,6,7,0,1,2].

Given the array nums after the possible rotation and an integer target,
return the index of target if it is in nums, or -1 if it is not in nums.

You must write an algorithm with O(log n) runtime complexity.

Constraints:

1 <= nums.length <= 5000
-104 <= nums[i] <= 104
All values of nums are unique.
nums is an ascending array that is possibly rotated.
-104 <= target <= 104
"""


def searchInRotated(nums, target):
    # binary search with a twist

    l, r = 0, len(nums)-1
    while l <= r:
        m = l + (r-l)//2
        print(f"{l=} {r=} {m=}")
        if nums[m] == target:
            return m
        print(f"{nums[l]} <= {target} and {target} < {nums[m]} = {nums[l] <= target and target < nums[m]}")
        # keep left half
        #  - left not rotated and target beteween nums[l] and nums[m]
        #  - left rotated but num is not in the right
        if (nums[l] < nums[m] and nums[l] <= target < nums[m]) \
           or (nums[l] > nums[m] and not (nums[m] < target <= nums[r])):
            r = m - 1
        # keep right half
        #  - all other cases
        else:
            l = m+1
    return -1



tests = [
    ([1,2,3,4,5], 7, -1),
    ([1,2,3,4,5], 4, 3),
    ([14,54,1,2,5,8,9], 5, 4),
    ([14,54,1,2,5,8,9], 54, 1),
    ([14,54,1,2,5,8,9], 14, 0),
    ([14,54,1,2,5,8,9], 1, 2),
    ([14,54,1,2,5,8,9], 9, 6),
    ([14,54,1,2,4,8,9], 5, -1),
    ([], 5, -1),
]
for t in tests:
    retVal = searchRotated(t[0], t[1])
    print(retVal, t)
    assert(retVal == t[2])
```

```python
# 76. Minimum Window Substring

def minWinSubstr(s, t):
    counter = {}
    for c in t:
        if c not in counter:
            counter[c] = 0
        counter[c] += 1

    print(f"{sum(counter.values())=}")
    
    minStr = ""
    minLen = len(s)
    left = 0
    right = -1

    # find first substring
    while right < len(s)-1 and sum(counter.values()) > 0:
        right += 1
        if s[right] in counter:
            counter[s[right]] -= 1

    if sum(counter.values()) > 0:
        return ""

    while left < right and s[left] not in counter:
        left += 1
    
    minStr = s[left:right+1]
    minLen = right - left + 1

    # print(f"Beginning: {minStr=} {minLen=}")
    
    while right < len(s)-1:
        while sum(counter.values()) == 0 or s[left] not in counter:
            if s[left] in counter:
                counter[s[left]] += 1
            left += 1

        # print(f"next: {left=} {right=} {s[left:right+1]}")
        
        while right < len(s)-1 and sum(counter.values()) > 0:
            right += 1
            if s[right] in counter:
                counter[s[right]] -= 1

        if sum(counter.values()) > 0:
            break

        if minLen > right - left + 1:
            minStr = s[left:right +1 ]
            minLen = right - left + 1

    return minStr

tests = [
    ("ADOBECODEBANC", "ABC", "BANC"),
    ("a", "a", "a"),
    ("a", "aa", ""),    
]
for t in tests:
    retVal = minWinSubstr(t[0], t[1])
    print(t, retVal)
    assert(t[2] == retVal)
```

```python
"""
https://leetcode.com/problems/symmetric-tree/

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

Constraints:

The number of nodes in the tree is in the range [1, 1000].
-100 <= Node.val <= 100
 

Follow up: Could you solve it both recursively and iteratively?
"""

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def leftIdx(idx):
    return idx * 2 + 1

def rightIdx(idx):
    return idx * 2 + 2

def readTree(vals):
    if len(vals) == 0:
        return None
    nodeMap = {}
    for i in range(len(vals)-1, -1, -1):
        if vals[i] == None:
            continue
        nodeMap[i] = TreeNode()
        nodeMap[i].val = vals[i]
        lidx = leftIdx(i)
        ridx = rightIdx(i)
        if lidx in nodeMap:
            nodeMap[i].left = nodeMap[lidx]
        if ridx in nodeMap:
            nodeMap[i].right = nodeMap[ridx]
    return nodeMap[0]

from collections import deque
def isSymmetric(root):
    if root == None:
        return True
    q = deque()
    q.append((root.left, root.right))
    while len(q) > 0:
        ln, rn = q.popleft()
        if ln == None and rn == None:
            continue
        if ln == None:
            return False
        if rn == None:
            return False
        if ln.val != rn.val:
            return False
        q.append((ln.left, rn.right))
        q.append((ln.right, rn.left))
    return True

tests = [
    ([1,2,2,3,4,4,3], True),
    ([1,2,2,None,3,None,3], False)
]
for t in tests:
    tree = readTree(t[0])
    assert(isSymmetric(tree) == t[1])
```

```python

from collections import defaultdict
def canFinishAllCourses(numCourses, prerequisites):
    """.
    
    directed graph / topological sorting

    [a, b]:   a -> b

        indegree(n): number of courses that have n as its prerequisite
        outdegree(n): number of courses that needs to be taken before n

    algorithm:
        try removing nodes that can be taken (outdegree==0) incrementally
        Adjust outdegree of nodes when removing a node n
        if we can remove all nodes, then return True
        otherwise, False
    """
    visited=set()
    outdegree = defaultdict(int)
    inNodes = defaultdict(list)

    for pre in prerequisites:
        a, b = pre[0], pre[1]
        outdegree[a]+= 1
        inNodes[b].append(a)

    toVisit = []
    for i in range(numCourses):
        if outdegree[i] == 0:
            toVisit.append(i)

    while len(toVisit):
        v = toVisit.pop()
        visited.add(v)
        for n in inNodes[v]:
            outdegree[n] -= 1
            if outdegree[n] == 0:
                toVisit.append(n)

    return len(visited) == numCourses

tests = [
    (2, [[1,0]], True),
    # There are a total of 2 courses to take. 
    # To take course 1 you should have finished course 0. So it is possible.
    (2, [[1,0],[0,1]], False),
    (5, [[0,1],
         [1,2],
         [2,3],
         [3,4],], True),
    (5, [[0,1],
         [1,2],
         [2,3],
         [3,4],
         [4,0]], False),
    (5, [[0,1],
         [2,1],
         [3,1],
         [4,1],], True),
    (5, [[0,1],
         [1,1],
         [2,1],
         [3,1],
         [4,1],], False),
]
for t in tests:
    retVal = canFinishAllCourses(t[0], t[1])
    print(t, retVal)
    assert(retVal == t[2])

```

```python
def minNumRooms(meetings):
    maxRooms = 0

    startTimes = []
    endTimes = []
    for s,e in meetings:
        startTimes.append(s)
        endTimes.append(e)

    startTimes.sort()
    endTimes.sort()

    numRooms = 0
    sIdx = 0
    eIdx = 0
    while sIdx < len(startTimes) or eIdx < len(endTimes):
        sVal = float('inf')
        eVal = float('inf')
        if sIdx < len(startTimes)-1:
            sVal = startTimes[sIdx]
        if eIdx < len(endTimes)-1:
            eVal = endTimes[eIdx]

        if sVal < eVal:
            numRooms += 1
            sIdx += 1
        elif sVal > eVal:
            numRooms -= 1
            eIdx += 1
        else: # sVal == eVal, one meeting ends, another meeting starts
            sIdx += 1
            eIdx += 1
        if numRooms > maxRooms:
            maxRooms = numRooms
    return maxRooms

tests = [
    ([[0, 30],[5, 10],[15, 20]], 2),
    ([[7,10],[2,4]], 1)
]

for t in tests:
    retVal = minNumRooms(t[0])
    print(t, retVal)
    assert(retVal == t[-1])
```

```python
import heapq

def minSpanningTree(n, connections):
    # keep a minheap with edge score
    # retrieve min score edge
    # if the edge connects merges separate groups of nodes, then use the edge
    # until all nodes are visited
    score = 0
    g = list(range(n))
    def find(x):
        if x != g[x]:
            g[x] = find(g[x])
        return g[x]

    minheap=[]
    for conn in connections:
        i,j,c = conn[0], conn[1], conn[2]
        heapq.heappush(minheap, [c, i, j])

    numEdges=0
    while len(minheap) > 0:
        conn = heapq.heappop(minheap)
        # print(f'{minheap=}, {conn=}, {score=}, {g=}')
        c,i,j = conn[0], conn[1], conn[2]
        if find(i-1) == find(j-1):
            continue
        g[find(j-1)] = find(i-1)
        score += c
        numEdges+=1
        if numEdges==n-1:
            # print(f'{minheap=}, {conn=}, {score=}, {g=}')
            return score

    return -1
```

```python

import threading
import time
from collections import deque

class BoundedBlockingQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = deque()
        self.producerSemaphore = threading.Semaphore(self.capacity)
        self.consumerSemaphore = threading.Semaphore(0)

    def enqueue(self, val):
        self.producerSemaphore.acquire()
        print("Produced", val)
        self.queue.append(val)
        self.consumerSemaphore.release()

    def dequeue(self):
        self.consumerSemaphore.acquire()
        val = self.queue.popleft()
        print("Consumed", val)
        self.producerSemaphore.release()
        return val

    def size(self):
        return len(self.queue)
```

```python

def prodOfArrExceptSelf(arr):
    N = len(arr)
    if N == 0:
        return []
    if N == 1:
        return [0]

    pLeft = [0]*N  #  prod of 0 .. i
    pRight = [0]*N #  prod of i .. N-1

    pL = 1
    pR = 1
    for i in range(N):
        pL *= arr[i]
        pLeft[i] = pL
        pR *= arr[N-1-i]
        pRight[N-1-i] = pR

    retVal = []
    for i in range(N):
        l = 1
        if i > 0:
            l = pLeft[i-1]
        r = 1
        if i < N-1:
            r = pRight[i+1]
        retVal.append(l*r)
    return retVal


# edge cases: arrval: 0, neg, arrlen: empty, 1, short, long
tests = [
    ([1,2,3,4], [24,12,8,6]),
    ([-1,1,0,-3,3], [0,0,9,0,0]),
    ([], []),
    ([12], [0]),
    ([-1], [0]),
    ([-5, 5], [5, -5]),
]

for t in tests:
    retVal = prodOfArrExceptSelf(t[0])
    print(t, retVal)
    assert(t[1] == retVal)
    
    
```
