```python
"""
https://leetcode.com/problems/meeting-rooms-ii/
subscription needed

https://leetcode.ca/all/253.html

Given an array of meeting time intervals consisting of start and end times [[s1,e1],[s2,e2],...] (si < ei),
find the minimum number of conference rooms required.

NOTE: input types have been changed on April 15, 2019. 
Please reset to default code definition to get new method signature.

"""

# sort the meetings
#         [          ]
#           [  ]
#            [    ]
#                [ ]
#         [ [[ ] []]]
# depth: 01 23 2 3210
# max:   01 23 3 3333
# this could be interpreted as () depth
# [          ]
#      [   ]
#   []
# [ [] [   ] ]
# same result as
# [        ] 
#      [     ]
#   []
# [ [] [   ] ]

# two solutions
#  1. sort meetings by start time
#     while processing meetings in order of start time:
#        find the earliest end time from already started meetings
#              ==>  keep a list of running meetings and retrieve the earliest end time

#  2. two sorted lists
#      start times
#        end times
#     compare the first items from the lists
#     if start time[0] < end time [0]
#         a new meeting room is needed
#         pop start time[0][
#     elif  start time[0] > end time[0]
#         a meeting room became available
#     else: # startime [0] == endtime[0]
#         a meeting ended, a meeting started at the same time (using the same room)
#         do nothing

import heapq
from collections import deque

def minNumRooms(meetings):
    if len(meetings) < 2:
        return len(meetings)

    # _meetings = sorted(meetings)
    # maxRooms = 0
    # runningMeetings = [] # minheap, sort by end time
    # for m in _meetings:
    #     # if the earliest ending meeting, if it ends before m starts, then pop
    #     # repeat until a meeting ends after m starts
    #     while len(runningMeetings)>0 and runningMeetings[0][0] < m[0]:
    #         heapq.heappop(runningMeetings)
    #     heapq.heappush(runningMeetings, (m[1], m))
    #     maxRooms = max(maxRooms, len(runningMeetings))

    # return maxRooms

    # # sorted lists of start and end times
    # startTs = []
    # endTs = []
    # for m in meetings:
    #     startTs.append(m[0])
    #     endTs.append(m[1])
    # startTs = deque(sorted(startTs))
    # endTs = deque(sorted(endTs))

    # maxNumRooms = 0
    # numRooms = 0
    # while len(startTs)>0 and len(endTs)>0:
    #     if startTs[0] < endTs[0]:
    #         numRooms += 1
    #         startTs.popleft()
    #         maxNumRooms = max(maxNumRooms, numRooms)
    #     elif startTs[0] > endTs[0]:
    #         numRooms -= 1
    #         endTs.popleft()
    #     else:
    #         # a meeting starts when another meeting ends
    #         startTs.popleft()
    #         endTs.popleft()
    # return maxNumRooms


    startTimes = []
    endTimes = []
    for s,e in meetings:
        startTimes.append(s)
        endTimes.append(e)

    startTimes.sort()
    endTimes.sort()

    maxRooms = 0
    numRooms = 0
    sIdx = 0
    eIdx = 0
    while sIdx < len(startTimes) or eIdx < len(endTimes):
        sVal = float('inf')
        eVal = float('inf')
        if sIdx < len(startTimes)-1:
            sVal = startTimes[sIdx]
        if eIdx < len(endTimes)-1:
            eVal = endTimes[eIdx]

        if sVal < eVal:
            numRooms += 1
            sIdx += 1
        elif sVal > eVal:
            numRooms -= 1
            eIdx += 1
        else: # sVal == eVal, one meeting ends, another meeting starts
            sIdx += 1
            eIdx += 1
        if numRooms > maxRooms:
            maxRooms = numRooms
    return maxRooms


tests = [
    ([[0, 30],[5, 10],[15, 20]], 2),
    ([[7,10],[2,4]], 1)
]
for t in tests:
    assert(minNumRooms(t[0]) == t[1])    
```
