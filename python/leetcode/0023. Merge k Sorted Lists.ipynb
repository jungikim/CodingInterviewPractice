```python
"""
https://leetcode.com/problems/merge-k-sorted-lists/


You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.
 

Constraints:

k == lists.length
0 <= k <= 10^4
0 <= lists[i].length <= 500
-10^4 <= lists[i][j] <= 10^4
lists[i] is sorted in ascending order.
The sum of lists[i].length will not exceed 104.
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def readLL(vals):
    if vals == None or len(vals)==0:
        return None
    dummy=ListNode()
    n = dummy
    for v in vals:
        n.next = ListNode(val=v)
        n = n.next
    return dummy.next

def dumpLL(ll):
    retVal = []
    n = ll
    while n:
        retVal.append(n.val)
        n = n.next
    return retVal

def printLL(ll):
    print(dumpLL(ll))

def readLLs(loVals):
    retVal=[]
    for vals in loVals:
        retVal.append(readLL(vals))
    return retVal

def dumpLLs(lol):
    retVal=[]
    for l in lol:
        retVal.append(dumpLL(l))
    return retVal

def printLLs(lol):
    for l in lol:
        printLL(l)


def mergeKSortedLists(kLinkedLists):
    if len(kLinkedLists) == 0:
        return None

# [1,4,5]
# [1,3,4]
# [2,6]

# 1. merge k at the same time
# find the smallest of the first values from all K lists
# put it in the merged list
# repeat
#
# 2. merge 1 and 2, merge 3,4 ... until all lists are merged  

    nodes = list(kLinkedLists)
    dummy = ListNode()
    merged = dummy

    while True:
        minIdx = -1
        for i, n in enumerate(nodes):
            if n == None:
                continue
            if minIdx == -1 or n.val < nodes[minIdx].val:
                minIdx = i
        if minIdx == -1:
            break
        merged.next = nodes[minIdx]
        merged = merged.next
        nodes[minIdx] = nodes[minIdx].next
    merged.next = None

    return dummy.next


tests = [
    ([[1,4,5],[1,3,4],[2,6]], [1,1,2,3,4,4,5,6]),
    ([], []),
    ([[]], [])
]
for t in tests:
    lol = readLLs(t[0])
    # printLLs(lol)
    merged = mergeKSortedLists(lol)
    # printLL(merged)
    # printLL(readLL(t[1]))
    assert(dumpLL(merged) == t[1])
```
