```python
from typing import List
import heapq

# # analytical?
# def maxNumber(nums1:List[int], nums2:List[int], k:int) -> List[int]:
# # conditions for the next number
# #  - we want the max number from either nums1 or nums2
# #    the picked number index should be low enough such that
# #    remaining k <= picked number + remaining num from that list + remaining nums from the other list

#     h1 = []
#     h2 = []
#     for i, n in enumerate(nums1):
#         heapq.heappush(h1, (-n,i))
#     for i, n in enumerate(nums2):
#         heapq.heappush(h2, (-n,i))

#     retVal=[]
#     while len(retVal)<k:
#         picked = -1
#         if len(h1)==0 or len(h2)==0:
#             pass # TODO
#         else:
#             max1, idx1 = h1[0]
#             max2, idx2 = h2[0]


#     return retVal


# bruteforce: generate all possible cases and save the max
# O( 2 ^ (M+N) )
# 4 moves:
#     skip 1
#     skip 2
#     pick 1
#     pick 2
def maxNumber(nums1:List[int], nums2:List[int], k:int) -> List[int]:
    _max = -1
    retVal = []
    def _skipOrPick(i1, i2):
        nonlocal _max
        if len(retVal)==k:
            v = int(''.join([str(i) for i in retVal]))
            if v > _max:
                _max = v
            # print(f'{i1=}, {i2=}, {v=}, {_max=}')
            return
        if _max > 0:
            for i,r in enumerate(retVal):
                if int(str(_max)[i]) < r:
                    break
                if int(str(_max)[i]) > r:
                    # print(f'not exploring {retVal=}, {i1=}, {i2=}, {_max=}')
                    return

        if i1 < len(nums1):
            _skipOrPick(i1+1, i2)
            retVal.append(nums1[i1])
            _skipOrPick(i1+1, i2)
            retVal.pop()
        if i2 < len(nums2):
            _skipOrPick(i1, i2+1)
            retVal.append(nums2[i2])
            _skipOrPick(i1, i2+1)
            retVal.pop()
    _skipOrPick(0, 0)
    return [int(c) for c in str(_max)]

tests = [
([3,4,6,5], [9,1,2,5,8,3], 5, [9,8,6,5,3]),
([6,7], [6,0,4], 5, [6,7,6,0,4]),
([3,9], [8,9], 3, [9,8,9]),
([3,3,3,3], [3,3,3,3], 4, [3,3,3,3]),
([3,3,4,3], [3,3,3,3], 4, [4,3,3,3]),
([3,3,3,4], [3,3], 4, [3,4,3,3]),
([1,1,1,4], [3,3], 4, [3,3,1,4]),
]
for t in tests:
    r = maxNumber(t[0], t[1], t[2])
    assert(r == t[3])
```
