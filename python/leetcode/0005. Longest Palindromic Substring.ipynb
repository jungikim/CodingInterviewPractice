```python
"""
https://leetcode.com/problems/longest-palindromic-substring/


Given a string s, return the longest palindromic substring in s.

Constraints:
1 <= s.length <= 1000
s consist of only digits and English letters.
"""


def longestPalindromicSubstr(s):
    if len(s) == 0:
        return ""
    # # bruteforce: O(n^3)
    # #   generate all possible substr
    # #   if isPalindromic and longer than max so far, then save
    # def _isPalindromic(s):
    #     if len(s) == 0: return False
    #     N = len(s)
    #     for i in range((N+1)//2):
    #         if s[i] != s[N-1-i]:
    #             return False
    #     return True
    # max = ""
    # for i in range(len(s)-1):
    #     for j in range(i, len(s)):
    #         substr = s[i:j+1]
    #         if _isPalindromic(substr) and len(substr)>len(max):
    #             max = substr
    # return max

    # O(n^2)
    max = s[0]
    for i in range(len(s)):
        # check for odd len palindrome
        offset=0
        while 0<=i-(offset+1) and i+(offset+1) < len(s) and s[i-(offset+1)]==s[i+(offset+1)]:
            offset+=1
        substr = s[i-offset:i+offset+1]
        if len(max) < len(substr) and len(substr) < len(s):
            max = substr
        #check for even len palindrome
        if i + 1 == len(s):
            continue
        if s[i] != s[i+1]:
            continue
        offset=0
        while 0<=i-(offset+1) and i+1+(offset+1) < len(s) and s[i-(offset+1)] == s[i+1+(offset+1)]:
            offset+=1
        substr = s[i-offset:i+1+offset+1]
        if len(max) < len(substr) and len(substr) < len(s):
            max = substr
    return max

tests = [
    ("babad", "bab"), # can also be "aba"
    ("cbbd", "bb")
]
for t in tests:
    assert(longestPalindromicSubstr(t[0]) == t[1])
```
