```python
"""
https://leetcode.com/problems/integer-to-roman/

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two one's added together.
12 is written as XII, which is simply X + II.
The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right.
However, the numeral for four is not IIII. Instead, the number four is written as IV.
Because the one is before the five we subtract it making four.
The same principle applies to the number nine, which is written as IX.

There are six instances where subtraction is used:
 - I can be placed before V (5) and X (10) to make 4 and 9. 
 - X can be placed before L (50) and C (100) to make 40 and 90. 
 - C can be placed before D (500) and M (1000) to make 400 and 900.
 - Given an integer, convert it to a roman numeral.

Constraints:
1 <= num <= 3999
"""

def intToRoman(num):
    """
    mapping for thousands, hundreds, tens, and ones
    """
    ones      = [ "", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX" ]
    tens      = [ "", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC" ]
    hundreds  = [ "", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM" ] 
    thousands = [ "", "M", "MM", "MMM", ]
    
    r1000 = thousands [ num // 1000 ]
    r100  = hundreds [ num // 100 % 10 ]
    r10   = tens [ num // 10 % 10 ]
    r1    = ones [ num // 1 % 10 ]

    return r1000 + r100 + r10 + r1


tests=[
    (1, "I"),
    (3, "III"),
    (58, "LVIII"),
    (1994, "MCMXCIV"),
    (3999, "MMMCMXCIX"),
]
for t in tests:
    retVal = intToRoman(t[0])
    print(t, retVal)
    assert(t[1] == retVal)
```

```python
"""
https://leetcode.com/problems/roman-to-integer/

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together.
12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right.
However, the numeral for four is not IIII. Instead, the number four is written as IV.
Because the one is before the five we subtract it making four.
The same principle applies to the number nine, which is written as IX.
There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer.

Q: input range?
empty input?
"""

def romanToInt(s):
    """
    lookahead needed for "subtract" cases
    I -> V X
    X -> L C
    C -> D M

    otherwise, left-to-right parsing?
    """

    vals = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50, 
        "C": 100,
        "D": 500,
        "M": 1000,
        #
        "IV": 4,
        "IX": 9,
        "XL": 40,
        "XC": 90,
        "CD": 400,
        "CM": 900,
    }
    
    if s is None or len(s) == 0:
        return f"Invalid Roman Numeral String: {s}"

    totalVal = 0
    idx = 0
    while idx < len(s):
        c = s[idx]
        if c not in vals:
            return f"Invalid Roman Numeral String: {s}"
        curVal = vals[ c ]
        if (c == "I" and idx + 1 < len(s) and (s[idx+1] == "V" or s[idx+1] == "X")) or \
           (c == "X" and idx + 1 < len(s) and (s[idx+1] == "L" or s[idx+1] == "C")) or \
           (c == "C" and idx + 1 < len(s) and (s[idx+1] == "D" or s[idx+1] == "M")):
            curVal = vals[ s[idx:idx+2] ]
            idx += 1

        idx += 1
        totalVal += curVal

    return totalVal

tests = [
    ("XII", 12),
    ("XXVII", 27),
    ("I", 1),
    ("III", 3),
    ("LVIII", 58),
    ("MCMXCIV", 1994),
    ("MMMCMXCIX", 3999),
]

for t in tests:
    retVal = romanToInt(t[0])
    print(t, retVal)
    assert(retVal == t[1])
```

```python
"""
https://leetcode.com/problems/3sum-closest/

Given an integer array nums of length n and an integer target, find three integers in nums such that the sum is closest to target.
Return the sum of the three integers.
You may assume that each input would have exactly one solution.

Constraints:
3 <= nums.length <= 500
-1000 <= nums[i] <= 1000
-104 <= target <= 104

Assuming that the numbers need to have unique idx from the list
(values can be same but should have different idx)
"""
def threeSumClosest(nums, target):
    """
    solution 1: loop over i, j, search for k (target-i-j), if not exact find the closest value
    """
    N = len(nums)
    if N == 0:
        return -1
    _sortedNums = sorted(nums)
    def _search(v):
        """
        returns closest value to v in _sortedNums
        """
        a, b = 0, N-1
        m = 0
        while a < b:
            m = a + (b-a)//2
            if _sortedNums[m] == v:
                return v
            if v < _sortedNums[m]:
                b = m-1
            else:
                a = m+1

        mindiff = float('inf')
        minidx = -1
        if 0<= m-1 < N and abs(v - _sortedNums[m-1]) < mindiff:
            mindiff = abs(v - _sortedNums[m-1])
            minidx = m-1
        if 0<= m < N and abs(v - _sortedNums[m]) < mindiff:
            mindiff = abs(v - _sortedNums[m])
            minidx = m
        if 0<= m+1 < N and abs(v - _sortedNums[m+1]) < mindiff:
            mindiff = abs(v - _sortedNums[m+1])
            minidx = m+1
        return minidx
            
    closest = float('inf')
    for i in range(N):
        for j in range(N):
            if i == j:
                continue
            kval = target - nums[i] - nums[j]
            k = _search(kval)
            if k == i or k==j:
                continue
            s = nums[i] + nums[j] + nums[k]
            if s < closest:
                closest = s
    return closest

tests = [
    ([], 0, -1),
    ([-1,2,1,-4], 1, 2),
    ([0,0,0], 1, 0)
]
for t in tests:
    retVal = threeSumClosest(t[0], t[1])
    print(t, retVal)
    assert(retVal == t[2])
```

```python
"""
https://leetcode.com/problems/letter-combinations-of-a-phone-number/

Given a string containing digits from 2-9 inclusive, 
return all possible letter combinations that the number could represent. 
Return the answer in any order.

A mapping of digits to letters (just like on the telephone buttons) is given below.
Note that 1 does not map to any letters.

Constraints:

0 <= digits.length <= 4
digits[i] is a digit in the range ['2', '9'].
"""

def letterCombinations(s):
    lettermap = {
        '2': ['a', 'b', 'c'],
        '3': ['d', 'e', 'f'],
        '4': ['g', 'h', 'i'],
        '5': ['j', 'k', 'l'],
        '6': ['m', 'n', 'o'],
        '7': ['p', 'q', 'r', 's'],
        '8': ['t', 'u', 'v'],
        '9': ['w', 'x', 'y', 'z'],
    }
    retVal = []
    if len(s) == 0:
        return retVal
    retVal.append("")
    for c in s:
        _retVal = []
        for r in retVal:
            for l in lettermap[c]:
                _retVal.append(r+l)
        retVal = _retVal
    return retVal

tests = [
    ("23", ["ad","ae","af","bd","be","bf","cd","ce","cf"]),
    ("", []),
    ("2", ["a","b","c"])
]
for t in tests:
    retVal = letterCombinations(t[0])
    print(t, retVal)
    assert(sorted(retVal) == sorted(t[1]))
```

```python
"""
https://leetcode.com/problems/valid-parentheses/

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', 
determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.

Constraints:

1 <= s.length <= 104
s consists of parentheses only '()[]{}'.

"""

def isValidParen(s):
    """
    state
    0: closed
    1: ( open
    2: { open
    3: [ open
    """
    state = 0
    for c in s:
        if c == '{' or c == '(' or c == '[':
            if state != 0:
                return False
            if c == '(':
                state = 1
            elif c == '{':
                state = 2
            else: # c == '[':
                state = 3
        else:
            if (c == ')' and state == 1 ) or \
               (c == '}' and state == 2 ) or \
               (c == ']' and state == 3 ):
                state = 0
            else:
                return False
    return True

tests = [
("()", True),
("()[]{}", True),
("(]", False)
]
for t in tests:
    retVal = isValidParen(t[0])
    print(t, retVal)
    assert(retVal == t[1])
```

```python
"""
https://leetcode.com/problems/merge-two-sorted-lists/

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list.
The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.


Constraints:
The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.
"""
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def readLL(vals):
    if len(vals) == 0:
        return None
    dummy=ListNode()
    n = dummy
    for v in vals:
        n.next = ListNode(val=v)
        n = n.next 
    return dummy.next

def dumpLL(ll):
    retVal = []
    n = ll
    while n:
        retVal.append(n.val)
        n = n.next
    return retVal

def printLL(ll):
    print(dumpLL(ll))

def mergeTwoSortedLists(l1, l2):
    dummy = ListNode(val=-1)
    p = dummy
    p1 = l1
    p2 = l2

    while p1 or p2:
        v1 = float('inf')
        v2 = float('inf')
        if p1:
            v1 = p1.val
        if p2:
            v2 = p2.val

        if v1 < v2:
            p.next = ListNode(v1)
            p1 = p1.next
        else:
            p.next = ListNode(v2)
            p2 = p2.next

        p = p.next

    return dummy.next

tests = [
    ([1,2,4], [1,3,4], [1,1,2,3,4,4]),
    ([], [], []),
    ([], [0], [0])
]
for t in tests:
    l1 = readLL(t[0])#; printLL(l1)
    l2 = readLL(t[1])#; printLL(l2)
    l3 = mergeTwoSortedLists(l1, l2)#; printLL(l3)
    assert(dumpLL(l3) == t[2])
```

```python
"""
https://leetcode.com/problems/generate-parentheses/

Given n pairs of parentheses, write a function to generate all combinations of well-formed parentheses.

Constraints:
1 <= n <= 8
"""

def genParen(n):
    """
    breath-first search
    (gen-so-far, level)
    level: 0 equal nums of ()
           1 one more open than close
           ...
            should never be less than 0
    """
    retVal = [("", 0)]
    for _ in range(n*2):
        _retVal = []
        for r in retVal:
            s,l = r
            if l > 0:
                _retVal.append((s+")", l-1))
            _retVal.append((s+"(", l+1))
        retVal = _retVal

    return [s for s,l in retVal if l == 0]

tests = [
    (3, ["((()))",
         "(()())",
         "(())()",
         "()(())",
         "()()()"]),
    (1, ["()"])
]
for t in tests:
    retVal = genParen(t[0])
    print(t, retVal)
    assert(sorted(retVal) == sorted(t[1]))
```

```python
"""
https://leetcode.com/problems/merge-k-sorted-lists/


You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.
 

Constraints:

k == lists.length
0 <= k <= 10^4
0 <= lists[i].length <= 500
-10^4 <= lists[i][j] <= 10^4
lists[i] is sorted in ascending order.
The sum of lists[i].length will not exceed 104.
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def readLL(vals):
    if vals == None or len(vals)==0:
        return None
    dummy=ListNode()
    n = dummy
    for v in vals:
        n.next = ListNode(val=v)
        n = n.next
    return dummy.next

def dumpLL(ll):
    retVal = []
    n = ll
    while n:
        retVal.append(n.val)
        n = n.next
    return retVal

def printLL(ll):
    print(dumpLL(ll))

def readLLs(loVals):
    retVal=[]
    for vals in loVals:
        retVal.append(readLL(vals))
    return retVal

def dumpLLs(lol):
    retVal=[]
    for l in lol:
        retVal.append(dumpLL(l))
    return retVal

def printLLs(lol):
    for l in lol:
        printLL(l)


def mergeKSortedLists(kLinkedLists):
    dummy = ListNode(val=-1)
    dummyP = dummy

    def _allNone(l):
        for i in l:
            if i is not None:
                return False
        return True

    K = len(kLinkedLists)
    pointers = list(kLinkedLists)

    while not _allNone(pointers):
        minVal = float('inf')
        minIdx = -1    
        for pi, p in enumerate(pointers):
            if p is None:
                continue
            if p.val < minVal:
                minIdx = pi
                minVal = p.val

        pointers[minIdx] = pointers[minIdx].next
        dummyP.next = ListNode(val=minVal)
        dummyP = dummyP.next

    return dummy.next
        


tests = [
    ([[1,4,5],[1,3,4],[2,6]], [1,1,2,3,4,4,5,6]),
    ([], []),
    ([[]], [])
]
for t in tests:
    retVal = mergeKSortedLists(readLLs(t[0]))
    print(t)
    printLL(retVal)
    print()
    assert(dumpLL(retVal) == t[1])
```

```python
"""
https://leetcode.com/problems/swap-nodes-in-pairs/

Given a linked list, swap every two adjacent nodes and return its head.
You must solve the problem without modifying the values in the list's nodes
(i.e., only nodes themselves may be changed.)
 

Constraints:
The number of nodes in the list is in the range [0, 100].
0 <= Node.val <= 100
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def toLL(l):
    sentinel = ListNode(val=-1,next=None)
    p = sentinel
    for i in l:
        p.next = ListNode(val=i)
        p = p.next
    return sentinel.next

def fromLL(head):
    retVal=[]
    p = head
    while p:
        retVal.append(p.val)
        p = p.next
    return retVal

def swapPairs(head):    
    p=head
    prev = None
    while p:
        a,b = None,None
        a = p
        if p.next:
            b = p.next
        if not a or not b:
            break
        c = b.next
        if prev == None:
            head = b
        else:
            prev.next = b
        b.next = a
        a.next = c
        p = c
        prev=a
    return head

tests = [
([1,2,3,4], [2,1,4,3]),
([], []),
([1], [1]),
]
for t in tests:
    head = toLL(t[0])
    retValLL = swapPairs(head)
    retVal = fromLL(retValLL)
    assert(t[1] == retVal)
```
