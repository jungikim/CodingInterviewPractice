```python
"""
https://leetcode.com/problems/symmetric-tree/

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

Constraints:

The number of nodes in the tree is in the range [1, 1000].
-100 <= Node.val <= 100
 

Follow up: Could you solve it both recursively and iteratively?
"""

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def leftIdx(idx):
    return idx * 2 + 1

def rightIdx(idx):
    return idx * 2 + 2

def readTree(vals):
    if len(vals) == 0:
        return None
    nodeMap = {}
    for i in range(len(vals)-1, -1, -1):
        if vals[i] == None:
            continue
        nodeMap[i] = TreeNode()
        nodeMap[i].val = vals[i]
        lidx = leftIdx(i)
        ridx = rightIdx(i)
        if lidx in nodeMap:
            nodeMap[i].left = nodeMap[lidx]
        if ridx in nodeMap:
            nodeMap[i].right = nodeMap[ridx]
    return nodeMap[0]

from collections import deque
def isSymmetric(root):
    if root == None:
        return True
    q = deque()
    q.append((root.left, root.right))
    while len(q) > 0:
        ln, rn = q.popleft()
        if ln == None and rn == None:
            continue
        if ln == None:
            return False
        if rn == None:
            return False
        if ln.val != rn.val:
            return False
        q.append((ln.left, rn.right))
        q.append((ln.right, rn.left))
    return True

tests = [
    ([1,2,2,3,4,4,3], True),
    ([1,2,2,None,3,None,3], False)
]
for t in tests:
    tree = readTree(t[0])
    assert(isSymmetric(tree) == t[1])
```
