```python
"""
https://leetcode.com/problems/palindrome-number/

Given an integer x, return true if x is a palindrome, and false otherwise.

Constraints:
-2^31 <= x <= 2^31 - 1
Follow up: Could you solve it without converting the integer to a string?
"""

def isPalindrome(x):
    if x < 0:
        return False
    if x < 10:
        return True

    s = str(x)
    l = len(s)
    for i in range(l//2):
        if s[i] != s[l-i-1]:
            return False
    return True

def isPalindromeInt(x):
    if x < 0:
        return False
    if x % 10 == 0:
        return False
    if x < 10:
        return True

    y = 0
    while y < x:
        y = 10 * y + x % 10
        x = x // 10
    return (x == y or x == y // 10)

tests = [
    (121, True),
    # 121 reads as 121 from left to right and from right to left.
    (-121, False),
    # From left to right, it reads -121.
    # From right to left, it becomes 121-.
    # Therefore it is not a palindrome.
    (10, False),
    # Reads 01 from right to left. Therefore it is not a palindrome.
    (12321, True),
    (-12321, False),
    (12344321, True),
    (99999199999, True)
]
for t in tests:
    retVal = isPalindrome(t[0])
    retValInt = isPalindromeInt(t[0])
    # print(t[0], t[1], retVal, retValInt)
    assert(retVal == t[1])
    assert(retValInt == t[1])
```
