```python
"""
https://leetcode.com/problems/non-overlapping-intervals/

Given an array of intervals intervals where intervals[i] = [starti, endi], 
return the minimum number of intervals you need to remove to make the rest of the intervals non-overlapping.


Constraints:
1 <= intervals.length <= 105
intervals[i].length == 2
-5 * 10^4 <= starti < endi <= 5 * 10^4
"""

# is this the same task as getting the maximum number of non-overlapping meetings?
# min_to_remove = total- max_can_keep


import heapq

def nonOverlapInts(intervals):
# sorted by start time, then by end time
#    1    2    3    4    5    6    7
#    XXXXX
#    XXXXXXXXXX
#         XXXXX
#              XXXXX

#    1    2    3    4    5    6    7
#    XXXXX
#    XXXXX
#    XXXXX

#    1    2    3    4    5    6    7
#    XXXXX
#         XXXXX

#    1    2    3    4    5    6    7
#    XXXXXXXXXXXXXXX
#         XXXXX
#              XXXXX

# we start to pick slots that end the earliest
# remove meetings that disqualifies (started or ended before the picked slot)

    minheap = []
    for interval in intervals:
        heapq.heappush(minheap, (interval[1], interval[0]))

    time = 0
    picked = []
    while len(minheap)>0:
        earliestToEnd = heapq.heappop(minheap)
        if time > earliestToEnd[1]:
            # print("start time in the past:", earliestToEnd)
            continue

        # print("Keep:", earliestToEnd)
        picked.append([earliestToEnd[1], earliestToEnd[0]])
        time = earliestToEnd[0]

    # print("Picked:", picked)
    numToRemove = len(intervals) - len(picked)
    return numToRemove



tests = [
    ([[1,2],[2,3],[3,4],[1,3]], 1),
    # remove [1,3]
    ([[1,2],[1,2],[1,2]], 2),
    # remove two [1,2]
    ([[1,2],[2,3]], 0)
]
for t in tests:
    assert(nonOverlapInts(t[0]) == t[1])
```
