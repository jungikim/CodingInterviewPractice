```python
"""
https://leetcode.com/problems/maximum-number-of-events-that-can-be-attended/

You are given an array of events where events[i] = [startDayi, endDayi]. Every event i starts at startDayi and ends at endDayi.

You can attend an event i at any day d where startTimei <= d <= endTimei. You can only attend one event at any time d.

Return the maximum number of events you can attend.


Constraints:

1 <= events.length <= 10^5
events[i].length == 2
1 <= startDayi <= endDayi <= 10^5
"""

# sort by start time
# 1 2
# 1 2 3 4
#   2 3
#     3 4

# sort by end time
# 1 2
#   2 3
# 1 2 3 4
#     3 4

# 1 2
#     3 4
# 1
#     3
# 1 2 3 4

# 1
# 1 2
# 1 2 3 4
#     3
#     3 4

# 1 2 3 4
#     3 4
#     3
# 1 2
# 1

# 
# day 1 .. N
  # consider/add events that start today
  # remove events that have past end dates, select one that has the earliest end date


import heapq

def maxAttend(events):
    if len(events) == 0:
        return 0
    # sort events by start time
    _events = sorted(events)
    startDay = _events[0][0]
    endDay = startDay
    for e in events:
        if endDay < e[1]:
            endDay = e[1]

    numAttended = 0
    addIdx = 0
    eventheap=[]
    while startDay <= endDay:
        # add events that start today
        while addIdx < len(_events) and _events[addIdx][0] <= startDay:
            heapq.heappush(eventheap, (_events[addIdx][1], _events[addIdx]))
            addIdx+=1
        # remove events in the past
        while len(eventheap)>0 and eventheap[0][0] < startDay:
            heapq.heappop(eventheap)
        # attend an event with the earliest end date
        if len(eventheap)>0:
            _, event = heapq.heappop(eventheap)
            numAttended += 1
        startDay += 1

    return numAttended

tests = [
    ([[1,2],[2,3],[3,4]], 3),
    ([[1,2],[2,3],[3,4],[1,2]], 4)
]
for t in tests:
    assert(maxAttend(t[0]) == t[1])
    
```
