```python
"""
https://leetcode.com/problems/n-queens/

The n-queens puzzle is the problem of placing n queens on an n x n chessboard 
such that no two queens attack each other.

Given an integer n, return all distinct solutions to the n-queens puzzle.
You may return the answer in any order.

Each solution contains a distinct board configuration of the n-queens' placement, 
where 'Q' and '.' both indicate a queen and an empty space, respectively.

Constraints:
1 <= n <= 9
"""

def solveNQueens(n):
    # analytical solution?
    # dfs with board, max n is 9 so do-able?
    # Try queen position in each row from 1..n
    # Next queen should be in the next row
    # function to check whether a position in the given row a queen can be placed
    #   or mark the board with a different char x to indicate the position is not suitable?
    #  x Q x x
    #  x x x Q
    #  . x x x
    #  . x . x

    board = [ ['.'] * n for _ in range(n) ]
    retVal = []

    def _search(r):
        if r == n:
            retVal.append([ ''.join(board[c]).replace('x','.') for c in range(n) ])
            # print('Success'); print('\n'.join([''.join(r) for r in board]))
            return
        for c in range(n):
            if board[r][c] != '.':
                continue
            # print('Try ', r, c)

            board[r][c] = 'Q'
            myBlocks = set()
            for d in [ (+1,0), (0, -1), (0,+1), (+1,-1), (+1,+1) ]:
                _r, _c = r, c
                while 0<= _r + d[0] < n and 0<= _c + d[1] < n:
                    _r, _c = _r + d[0], _c + d[1]
                    if board[_r][_c] == '.':
                        board[_r][_c] = 'x'
                        myBlocks.add((_r,_c))
            # print(r); print('\n'.join([''.join(r) for r in board]))
            _search(r+1)
            board[r][c] = '.'
            for (_r,_c) in myBlocks:
                board[_r][_c] = '.'

    _search(0)

    return retVal


tests = [
    (4, [[".Q..",
          "...Q",
          "Q...",
          "..Q."],
         ["..Q.",
          "Q...",
          "...Q",
          ".Q.."]]),
     # There exist two distinct solutions to the 4-queens puzzle as shown above
     (1, [["Q"]]),
     # (9, []),
]
for t in tests:
    retVal = solveNQueens(t[0])
    # for i,board in enumerate(retVal):
    #     print(i+1)
    #     print('\n'.join([''.join(r) for r in board]))
    assert(retVal == t[1])
```
