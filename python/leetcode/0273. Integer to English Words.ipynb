```python
"""
https://leetcode.com/problems/integer-to-english-words/


Convert a non-negative integer num to its English words representation.

 
Constraints:

0 <= num <= 2^31 - 1
"""

def numToEnglish(num):
    # we want to independently process numbers thounsands
    # 123,456,789
    # ___ ___ ___
    # 123 million 456 thousands 789
    #

    # max num is 2^31-1, which is 2147483647
    # 2 147 483 647
    # so we need "Billion", "Million", "Thousand" "" multipliers

    multipliers = ["", "Thousand", "Million", "Billion"]
    ones = ["", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", ]
    teens = ["Ten", "Eleven", "Twelve", "Thirteen", "Fourteen", "Fifteen", "Sixteen", "Seventeen", "Eighteen", "Nineteen", ]
    tens = ["", "", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"]
    hundreds = [""] + [ one + " Hundred" for one in ones[1:] ]

    retVal = ""
    n = num
    mLevel = 0
    while n > 0:
        readout = n % 1000 
        read_100 = hundreds [ readout // 100 ]
        read_10 = ""
        read_1 = ""
        if readout % 100 >= 20: 
            read_10 = tens[readout % 100 // 10]
            read_1 = ones[readout % 10]
        else:
            read_10 = (ones+teens) [ readout % 20 ]

        read = read_100 + " " + read_10 + " " + read_1 + " " + multipliers[mLevel]
        mLevel += 1
        n = n // 1000
        retVal = read + " " + retVal

    if len(retVal) == 0:
        retVal = "Zero"
    
    retVal = retVal.strip().replace("  ", " ")

    # print(retVal)
    return retVal


tests = [
    (0, "Zero"),
    (12000, "Twelve Thousand"),
    (120, "One Hundred Twenty"),
    (123, "One Hundred Twenty Three"),
    (12345, "Twelve Thousand Three Hundred Forty Five"),
    (1234567, "One Million Two Hundred Thirty Four Thousand Five Hundred Sixty Seven")
]
for t in tests:
    assert(numToEnglish(t[0]) == t[1])
```
