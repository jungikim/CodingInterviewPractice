```python
"""
https://leetcode.com/problems/reorder-data-in-log-files/


You are given an array of logs.
Each log is a space-delimited string of words, where the first word is the identifier.

There are two types of logs:

Letter-logs: All words (except the identifier) consist of lowercase English letters.
Digit-logs: All words (except the identifier) consist of digits.

Reorder these logs so that:
The letter-logs come before all digit-logs.
The letter-logs are sorted lexicographically by their contents. 
If their contents are the same, then sort them lexicographically by their identifiers.
The digit-logs maintain their relative ordering.
Return the final order of the logs.


Constraints:

1 <= logs.length <= 100
3 <= logs[i].length <= 100
All the tokens of logs[i] are separated by a single space.
logs[i] is guaranteed to have an identifier and at least one word after the identifier.
"""

def reorderDataInLogFiles(logs):
    if len(logs) == 0:
        return logs

    # letterLogs (sorted lexicographically by the contents, then identifier) + digitLogs (reserve order)
    letterLogs = []
    digitLogs = []

    def _split(log):
        if len(log)==0:
            return "", ""
        cIdx = 0
        while cIdx < len(log) and log[cIdx] != ' ':
            cIdx += 1
        id = log[:cIdx]
        content = log[cIdx+1:]
        return id, content
    
    def _isDigitLog(content):
        cIdx = 0
        isAllDigits = True
        while cIdx < len(content):
            if content[cIdx] == ' ' or content[cIdx].isdigit():
                cIdx += 1
                continue
            isAllDigits = False
            cIdx += 1
        return isAllDigits

    # def _isLetterLog(content):
    #     return not _isDigitLog(content)

    for l in logs:
        id, content = _split(l)
        if _isDigitLog(content):
            digitLogs.append(l)
        else:
            letterLogs.append((content, id))

    letterLogs = sorted(letterLogs)
    letterLogs = [ id+" "+content for content, id in letterLogs ]
    
    return letterLogs + digitLogs


tests = [
    (["dig1 8 1 5 1",
      "let1 art can",
      "dig2 3 6",
      "let2 own kit dig",
      "let3 art zero"],
    ["let1 art can",
     "let3 art zero",
     "let2 own kit dig",
     "dig1 8 1 5 1",
     "dig2 3 6"]),
    #The letter-log contents are all different, so their ordering is "art can", "art zero", "own kit dig".
    #The digit-logs have a relative order of "dig1 8 1 5 1", "dig2 3 6".
    (["a1 9 2 3 1",
      "g1 act car",
      "zo4 4 7",
      "ab1 off key dog",
      "a8 act zoo"],
    ["g1 act car",
     "a8 act zoo",
     "ab1 off key dog",
     "a1 9 2 3 1",
     "zo4 4 7"])
]

for t in tests:
    assert(reorderDataInLogFiles(t[0]) == t[1])
```
