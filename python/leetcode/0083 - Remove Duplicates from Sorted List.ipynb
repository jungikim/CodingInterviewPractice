```python
"""
https://leetcode.com/problems/remove-duplicates-from-sorted-list/

Given the head of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list sorted as well.

Constraints:
The number of nodes in the list is in the range [0, 300].
-100 <= Node.val <= 100
The list is guaranteed to be sorted in ascending order.
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def toLL(l):
    sentinel = ListNode(val=-1)
    p = sentinel
    for i in l:
        p.next = ListNode(val=i)
        p = p.next
    return sentinel.next

def toList(head):
    p = head
    retVal = []
    while p:
        retVal.append(p.val)
        p = p.next
    return retVal

from typing import Optional

def deleteDuplicates(head: Optional[ListNode]) -> Optional[ListNode]:
    sentinel = ListNode(val=-1)
    p = sentinel
    i = head
    while i:
        if i and i.next:
            while i and i.next and i.val == i.next.val:
                i = i.next
            # advance i until i is the last of the repeated
        p.next = i
        p = p.next
        i = i.next
    p.next = None
    return sentinel.next

tests = [
    ([1,1,2], [1,2]),
    ([1,1,2,3,3], [1,2,3]),
    ([], []),
    ([1,1], [1]),
    ([1,1,2,4,5,5], [1,2,4,5]),
]
for t in tests:
    ll = toLL(t[0])
    retVal = deleteDuplicates(ll)
    retValList = toList(retVal)
    # print(t, retValList)
    assert(retValList == t[1])
```
