```python
"""
https://leetcode.com/problems/3sum-closest/

Given an integer array nums of length n and an integer target, find three integers in nums such that the sum is closest to target.
Return the sum of the three integers.
You may assume that each input would have exactly one solution.

Constraints:
3 <= nums.length <= 500
-1000 <= nums[i] <= 1000
-104 <= target <= 104

Assuming that the numbers need to have unique idx from the list
(values can be same but should have different idx)
"""

def threesumclosest(nums, target):
    sum = 0

    # # brute force O(n^3)
    # # all combinations
    # mindiff = pow(2,31)-1
    # for i in range(0, len(nums)-2):
    #     for j in range(i+1, len(nums)-1):
    #         for k in range(j+1, len(nums)):
    #             s = (nums[i]+nums[j]+nums[k])
    #             diff = abs(target-s)
    #             if  mindiff > diff:
    #                 mindiff = diff
    #                 sum = s

    #    can sorting the numbers help?
    #     for two numbers and finding min diff from target, 
    #        [ 0 1 2 3 4 5 6 7 ]      5
    #          lp            rp
    #          sum = 7, 2 over the target (update global), decrease sum (rp--)
    #          lp         rp
    #          sum = 6, 1 over the target (update global), decrease sum (rp--)
    #          lp       rp
    #          sum = 5, same as target (update global), return sum
    #        update global
    #     for the third number, we can add an outer loop 
    #    O(n^2)
    snums = sorted(nums)
    mindiff = pow(2,31)-1
    for i in range(len(snums)-2):
        lp = i+1
        rp = len(snums)-1
        while lp < rp:
            s = snums[i] + snums[lp] + snums[rp]
            diff = s - target
            adiff = abs(diff)
            if diff == 0:
                return s
            if  mindiff > adiff:
                mindiff = adiff
                sum = s
            if diff > 0:
                rp -= 1
            else:
                lp += 1

    return sum


tests=[
    ([-1,2,1,-4], 1, 2),
    ([0,0,0], 1, 0)
]
for t in tests:
    assert(threesumclosest(t[0], t[1]) ==  t[2])
```
