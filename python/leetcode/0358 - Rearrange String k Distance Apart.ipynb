```python
"""
https://leetcode.com/problems/rearrange-string-k-distance-apart/description/
https://leetcode.ca/2016-11-22-358-Rearrange-String-k-Distance-Apart/


Given a string s and an integer k, rearrange s such that the same characters are at least distance k from each other.
If it is not possible to rearrange the string, return an empty string "".


Constraints:
1 <= s.length <= 3 * 10^5
s consists of only lowercase English letters.
0 <= k <= s.length
"""

from collections import defaultdict
import heapq

def kDistApart(s, k):
    l = len(s)
    maxFreq = 0
    counter = defaultdict(int)
    for c in s:
        counter[c] += 1
        maxFreq = max(maxFreq, counter[c])
    def _isPossible():
        for ch, cnt in counter.items():
            if (l - cnt) < (cnt-1)*(k-1):
                return False
        return True
    if not _isPossible():
        return ""

    heap = []
    for ch, cnt in counter.items():
        heapq.heappush(heap, (-cnt, ch))

    retVal=[]
    while len(retVal) < len(s):
        addback = []
        for i in range(k):
            if len(heap)==0:
                break
            cnt, ch = heapq.heappop(heap)
            retVal.append(ch)
            addback.append((ch, cnt+1))
        for (ch, cnt) in addback:
            if cnt == 0: continue
            heapq.heappush(heap, (cnt, ch))
    return ''.join(retVal)

tests = [
    ("aabbcc", 3, "abcabc"),
    ("aaabc", 3, ""),
    ("aabc", 3, "abca"),
    ("aaabcd", 3, ""),
    ("aaabcde", 3, "abcadea"),
    ("aaadbbcc", 2, "abacabcd"),
]
for t in tests:
    retVal = kDistApart(t[0], t[1])
    assert(retVal == t[2])
```
