```python
"""
https://leetcode.com/problems/container-with-most-water/

You are given an integer array height of length n.
There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.
"""

def mostwater(heights):
    mostwater = 0
    # # bruteforce
    # for i in range(len(heights)-1):
    #     for j in range(i+1, len(heights)):
    #         base = j-i
    #         height = min(heights[i], heights[j])
    #         water = base * height
    #         if mostwater < water:
    #             mostwater = water

    # dynamic window: O(n)
    #   not quite intuitive,
    #     counter example would be where
    #          discarded height (lower of the two) may be used in an optimal case, or
    #          once pointers move left-right, the actual optimal case is 
    #              entirely to the left or right of the left-right window
    #     these cases are not actually possible:
    #           lower of the two would be never be used in a better case
    #           since left of the left pointer are all lower than left p, and vice versa for right,
    #           the optimal case cannot be entirely to the left or the right of the window
    left = 0
    right = len(heights)-1
    while left < right:
        base = right - left
        height = min(heights[left], heights[right])
        water = base * height
        if mostwater < water:
            mostwater = water
        if heights[left] <= heights[right]:
            left += 1
        else:
            right -= 1

    return mostwater

    
tests = [
([1,8,6,2,5,4,8,3,7], 49),
([1,1], 1)
]
for t in tests:
    assert(mostwater(t[0])==t[1])
```
