```python
"""
https://leetcode.com/problems/reverse-linked-list/

Given the head of a singly linked list, reverse the list, and return the reversed list.


Constraints:

The number of nodes in the list is the range [0, 5000].
-5000 <= Node.val <= 5000
 
Follow up: A linked list can be reversed either iteratively or recursively. Could you implement both?
"""




class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def readLL(vals):
    dummy=ListNode()
    n = dummy
    for v in vals:
        n.next = ListNode(val=v)
        n = n.next
    return dummy.next

def dumpLL(ll):
    retVal = []
    n = ll
    while n:
        retVal.append(n.val)
        n = n.next
    return retVal

def printLL(ll):
    print(dumpLL(ll))


def reverseList(ll):
    stack=[]
    n = ll
    while n:
        stack.append(n)
        n = n.next

    dummy=ListNode()
    n = dummy
    while len(stack):
        nextNode = stack.pop()
        n.next = nextNode
        n = n.next
    n.next = None

    return dummy.next
    

tests = [
    ([1,2,3,4,5], [5,4,3,2,1]),
    ([1,2], [2,1]),
    ([], [])
]
for t in tests:
    ll = readLL(t[0])
    # printLL(ll)
    nl = reverseList(ll)
    # printLL(nl)
    assert(dumpLL(nl) == dumpLL(readLL(t[1])))
```
