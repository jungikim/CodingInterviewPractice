```python
"""
https://leetcode.com/problems/rotate-list/description/


Given the head of a linked list, rotate the list to the right by k places.

Constraints:
The number of nodes in the list is in the range [0, 500].
-100 <= Node.val <= 100
0 <= k <= 2 * 10^9
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def readLL(arr):
    sentinelHead = ListNode(val=-1,next=None)
    p = sentinelHead
    for n in arr:
        p.next = ListNode()
        p = p.next
        p.val = n
    return sentinelHead.next

def fromLL(head):
    retVal = []
    p = head
    while p:
        retVal.append(p.val)
        p = p.next
    return retVal

def rotateRight(head, k):
    N = 0
    p = head
    while p:
        p = p.next
        N+=1
    if N == 0:
        return head
    _k = k % N
    if _k == 0:
        return head
    p = head
    for _ in range(N-_k-1):
        p = p.next

    newHead = p.next
    t = p
    for _ in range(_k):
        t = t.next
    t.next = head
    p.next = None
    
    return newHead



tests = [
    ([1,2,3,4,5], 2, [4,5,1,2,3]),
    ([0,1,2], 4, [2,0,1]),
]
for t in tests:
    l = readLL(t[0])
    retVal = rotateRight(l, t[1])
    ll = fromLL(retVal)
    # print(t, ll)
    assert(t[2] == ll)
```
