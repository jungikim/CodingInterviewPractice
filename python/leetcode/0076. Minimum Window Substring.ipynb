```python
"""
https://leetcode.com/problems/minimum-window-substring/

Given two strings s and t of lengths m and n respectively,
return the minimum window substring of s such that every character in t (including duplicates) is included in the window.
If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.

Constraints:

m == s.length
n == t.length
1 <= m, n <= 105
s and t consist of uppercase and lowercase English letters.
 

Follow up: Could you find an algorithm that runs in O(m + n) time?
"""

# same as the one I already solved some days ago, but trying again
# and in just a few days, I already forget how I solved this...
# after thought: very difficult and tricky; should try again later

from collections import defaultdict

def minWinSubstr(s, t):
    retVal = ""
    if len(s) == 0 or len(s) < len(t):
        return retVal

    counter = defaultdict(int)
    for c in t:
        counter[c] += 1

    # s[left:right]
    left = 0
    right = 0 # will be incremented as soon as we enter the loop
    while right < len(s):
        # increase right by 1; mandatory to keep the while loop going
        if s[right] in counter:
            counter[s[right]] -= 1
        right+=1
        # increase right until the requirement is met
        while right < len(s) and \
              sum([c for c in counter.values() if c > 0]) > 0:
            if s[right] in counter:
                counter[s[right]] -= 1
            right+=1
        # decrease left as long as we meet the requirmenet
        while left < len(s) and \
                (s[left] not in counter or counter[s[left]] < 0):
            if s[left] in counter:
                counter[s[left]] += 1
            left+=1
        # if requirment is met and we have a better solution, then save
        if sum([c for c in counter.values() if c > 0]) <= 0:
            if retVal == "" or len(s[left:right]) < len(retVal):
                retVal = s[left:right]

    return retVal


tests = [
    ("ADOBECODEBANC", "ABC", "BANC"),
    ("a", "a", "a"),
    ("a", "aa", "")
]
for t in tests:
    assert(minWinSubstr(t[0], t[1]) == t[2])
```
