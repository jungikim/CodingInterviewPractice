```python
"""
https://leetcode.com/problems/binary-tree-zigzag-level-order-traversal/

Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. 
(i.e., from left to right, then right to left for the next level and alternate between).

Constraints:

The number of nodes in the tree is in the range [0, 2000].
-100 <= Node.val <= 100
"""

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def leftidx(idx):
    return idx * 2 + 1

def rightidx(idx):
    return idx * 2 + 2

def readTree(vals):
    if len(vals)==0:
        return None
    nodeMap = {}
    for i in range(len(vals)-1, -1, -1):
        li, ri = leftidx(i), rightidx(i)
        nodeMap[i] = TreeNode(val=vals[i],
                              left=nodeMap[li] if li in nodeMap else None,
                              right=nodeMap[ri] if ri in nodeMap else None)
    return nodeMap[0]

from collections import deque
def zigzag(root):
    retVal = []
    if root == None:
        return retVal

    q = deque()
    q.
    q.append(root)
    # level-by-level breath first
    while len(q)>0:
        level = len(retVal)
        curLevel = []
        nextLevel = deque()
        while len(q) > 0:
            n = q.popleft()
            if n.val:
                curLevel.append(n.val)
            if n.left:
                nextLevel.append(n.left)
            if n.right:
                nextLevel.append(n.right)
        if level % 2 == 1:
            curLevel.reverse()
        retVal.append(curLevel)
        q = nextLevel

    return retVal

tests = [
    ([3,9,20,None,None,15,7], [[3],[20,9],[15,7]]),
    ([1], [[1]]),
    ([], [])
]
for t in tests:
    tree=readTree(t[0])
    assert(zigzag(tree) == t[1])
```
