```python
"""
https://leetcode.com/problems/maximum-average-subtree/
https://leetcode.ca/2018-12-24-1120-Maximum-Average-Subtree/

Given the root of a binary tree, return the maximum average value of a subtree of that tree. Answers within 10-5 of the actual answer will be accepted.
A subtree of a tree is any node of that tree plus all its descendants.
The average value of a tree is the sum of its values, divided by the number of nodes.


Constraints:
The number of nodes in the tree is in the range [1, 10^4].
0 <= Node.val <= 10^5
"""

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def readTree(nums):
    nodeMap={}
    for i in range(len(nums)-1, -1, -1):
        if nums[i] == None:
            continue
        nodeMap[i] = TreeNode(val=nums[i],
                              left=nodeMap[i*2+1] if i*2+1 in nodeMap else None,
                              right=nodeMap[i*2+2] if i*2+2 in nodeMap else None)
    return nodeMap[0]


def maxAvgSubtree(root):
    maxAvg = 0
    def _search(node):
        if node == None:
            return 0, 0 # sum, cnt
        lsum, lcnt = _search(node.left)
        rsum, rcnt = _search(node.right)
        cnt = lcnt + rcnt + 1
        _sum = lsum + rsum + node.val
        nonlocal maxAvg
        maxAvg = max(maxAvg, _sum/cnt)
        return _sum, cnt

    _search(root)
    return maxAvg


tests = [
([5,6,1], 6.00000),
# For the node with value = 5 we have an average of (5 + 6 + 1) / 3 = 4.
# For the node with value = 6 we have an average of 6 / 1 = 6.
# For the node with value = 1 we have an average of 1 / 1 = 1.
# So the answer is 6 which is the maximum.
([0,None,1], 1.00000)
]
for t in tests:
    root=readTree(t[0])
    assert(maxAvgSubtree(root) == t[1])
```
