```python
"""
https://leetcode.com/problems/merge-two-sorted-lists/

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list.
The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.


Constraints:
The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.
"""


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def readLL(vals):
    if len(vals) == 0:
        return None
    dummy=ListNode()
    n = dummy
    for v in vals:
        n.next = ListNode(val=v)
        n = n.next 
    return dummy.next


def dumpLL(ll):
    retVal = []
    n = ll
    while n:
        retVal.append(n.val)
        n = n.next
    return retVal


def printLL(ll):
    print(dumpLL(ll))


def mergeTwoLists(list1, list2):
    dummy = ListNode()
    n1 = list1
    n2 = list2
    n3 = dummy
    while n1 or n2:
        if n1 and n2:
            v1 = n1.val
            v2 = n2.val
            if v1 < v2:
                n3.next = n1
                n3 = n3.next
                n1 = n1.next
            else:
                n3.next = n2
                n3 = n3.next
                n2 = n2.next

        elif n1 and not n2:
            n3.next = n1
            break

        elif n2 and not n1:
            n3.next = n2
            break

    return dummy.next


tests = [
    ([1,2,4], [1,3,4], [1,1,2,3,4,4]),
    ([], [], []),
    ([], [0], [0])
]
for t in tests:
    l1 = readLL(t[0])#; printLL(l1)
    l2 = readLL(t[1])#; printLL(l2)
    l3 = mergeTwoLists(l1, l2)#; printLL(l3)
    assert(dumpLL(l3) == t[2])
```
