```python
"""
Given the head of a linked list and a value x, 
partition it such that all nodes less than x come before nodes greater than or equal to x.

You should preserve the original relative order of the nodes in each of the two partitions.

 
Example 1:


Input: head = [1,4,3,2,5,2], x = 3
Output: [1,2,2,4,3,5]

Example 2:
Input: head = [2,1], x = 2
Output: [1,2]
 
Constraints:
The number of nodes in the list is in the range [0, 200].
-100 <= Node.val <= 100
-200 <= x <= 200
"""

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def arr2LL(arr):
    dummy = ListNode()
    p = dummy
    for n in arr:
        p.next = ListNode(val=n)
        p = p.next
    return dummy.next

def LL2Arr(ll):
    retVal = []
    p = ll
    while p:
        retVal.append(p.val)
        p = p.next
    return retVal


def printLL(ll):
    p = ll
    while p:
        print(p.val, end=",")
        p = p.next
    print()

def partition(head, x):
    # sentinel nodes
    lessthan = ListNode()
    greaterthanequalto = ListNode()

    lessthanP = lessthan
    greatthanequaltoP = greaterthanequalto
    p = head
    while p:
        if p.val < x:
            lessthanP.next = p
            lessthanP = lessthanP.next
        else:
            greatthanequaltoP.next = p
            greatthanequaltoP = greatthanequaltoP.next
        p = p.next

    greatthanequaltoP.next = None
    lessthanP.next = greaterthanequalto.next
    return lessthan.next


tests = [
    ([1,4,3,2,5,2], 3, [1,2,2,4,3,5]),
    ([2,1], 2, [1,2]),
]

for t in tests:
    l1 = arr2LL(t[0])
    # printLL(l1)
    ans = partition(l1, t[1])
    # printLL(ans)
    ansArr = LL2Arr(ans)
    assert(ansArr == t[2])
```
