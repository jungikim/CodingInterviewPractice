```python
"""
https://leetcode.com/problems/validate-binary-search-tree/


Given the root of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:

The left subtree of a node contains only nodes with keys less than the node's key.
The right subtree of a node contains only nodes with keys greater than the node's key.
Both the left and right subtrees must also be binary search trees.

Constraints:

The number of nodes in the tree is in the range [1, 104].
-231 <= Node.val <= 231 - 1
"""

from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def leftIdx(idx):
    return idx * 2 + 1

def rightIdx(idx):
    return idx * 2 + 2

def readTree(vals):
    if len(vals) == 0:
        return None
    nodeMap={}
    for i in range(len(vals)-1, -1, -1):
        if vals[i] == None:
            continue
        nodeMap[i] = TreeNode()
        nodeMap[i].val = vals[i]
        lidx = leftIdx(i)
        ridx = rightIdx(i)
        if lidx in nodeMap:
            nodeMap[i].left = nodeMap[lidx]
        if ridx in nodeMap:
            nodeMap[i].right = nodeMap[ridx]
    return nodeMap[0]

def printTree(node, depth):
    if node == None:
        return
    printTree(node.right, depth+1)
    print('\t'*depth, node.val)
    printTree(node.left, depth+1)
    
def isValidBTree(root):
    if root == None:
        return True

    # # recursive version
    # def _isValBTree(n, lower, upper):
    #     if n == None:
    #         return True
    #     if n.left and n.left.val < lower:
    #         return False
    #     if n.right and upper < n.right.val:
    #         return False        
    #     return _isValBTree(n.left, lower, n.val) and \
    #            _isValBTree(n.right, n.val, upper)
    
    # return _isValBTree(root, float('-inf'), float('inf'))

    # DFS version
    stack = []
    stack.append((root, float('-inf'), float('inf')))
    while len(stack)>0:
        n, lowerb, upperb = stack.pop()
        if n.left and n.left.val < lowerb:
            return False
        if n.right and n.right.val > upperb:
            return False        
        if n.left:
            stack.append((n.left, lowerb, n.val))
        if n.right:
            stack.append((n.right, n.val, upperb))
    return True

tests = [
    ([2,1,3], True),
    ([5,1,4,None,None,3,6], False)
]
for t in tests:
    root = readTree(t[0])
    # printTree(root, 0)
    # print()
    assert(isValidBTree(root) == t[1])
```
